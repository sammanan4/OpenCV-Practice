<h3>Canny Edge Detection </h3>

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  Read the images
img = cv.imread('DATA/sammy_face.jpg')
plt.imshow(img)
```

```python
# If we straightout perform edge detection we will detect too many edges due to noise
# Let's see
# we will first try to evaluate two thresholds based on the median of the images instead of brute guessing
medianVal = np.median(img)
lower = int(max(0, 0.7 * medianVal))
upper = int(min(255, 1.8 * medianVal))
```

```python
edges = cv.Canny(img, threshold1 = lower+50, threshold2 = upper+100)
```

```python
plt.imshow(edges, cmap='gray')
```

<h3> Let's apply blurring and then try edge detection </h3>

```python
blurredImg = cv.medianBlur(img, 7)
```

```python
edges = cv.Canny(blurredImg, threshold1 = lower, threshold2 = upper+20)
```

```python
plt.imshow(edges, cmap='gray')
```
