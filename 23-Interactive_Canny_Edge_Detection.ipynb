```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact,interactive,fixed
import ipywidgets as widgets

#  Read the images
img = cv.imread('C://Users/Manan/PythonComputerVision/DATA/sammy_face.jpg')
plt.imshow(img)

blurredImg = cv.medianBlur(img, 7)
```

```python
plt.imshow(blurredImg)
```

```python
def applyCanny(thresh1, thresh2):
    edges = cv.Canny(blurredImg, threshold1 = thresh1, threshold2 = thresh2)
    plt.imshow(edges, cmap='gray')
interact(applyCanny, thresh1 = widgets.IntSlider(min = 0, max = 255, step = 1, value = 0), thresh2 = widgets.IntSlider(min = 0, max = 255, step = 1, value = 0))
```
