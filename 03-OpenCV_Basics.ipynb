<h3> We can see that the image above has incorrect color scheme </h3><br>
The reason is that matplotlib expects the image to have channel sequence as RGB,
whereas OpenCV works with BGR (which is the reverse of matplotlib).
<br><br>
So the color values for blue are interpreted as red by matplotlib, hence the incorrect color mapping


```python
# we can manually do something like
actual_pic = pic.copy()
actual_pic[:,:,0] = pic[:,:,2]
actual_pic[:,:,2] = pic[:,:,0]
plt.imshow(actual_pic)
```

<h4> However the hardwork is not necessary as we already have a function to do this reversal </h4>

```python
actual_pic = cv.cvtColor(pic, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(actual_pic)
```

```python
# we can also choose many other color conversions
gray_pic = cv.cvtColor(pic, cv.COLOR_BGR2GRAY)
# or
# gray_pic = cv.cvtColor(pic, cv.IMREAD_GRAYSCALE)
```

```python
plt.imshow(gray_pic)
```

```python
# for matplotlib's converntions we have to use color mapping
plt.imshow(gray_pic, cmap = 'gray')
```

```python
gray_pic.shape
```

<h3> We can flip the images </h3>

```python
plt.imshow(cv.flip(actual_pic, 0))
```

<h3> We can save an image </h3><br>
The color needs to be converted back to  BGR 

```python
save_pic = cv.cvtColor(cv.flip(actual_pic, 0), cv.COLOR_RGB2BGR)
```

```python
cv.imwrite('Flipped_Dog.jpg', save_pic) 
```
