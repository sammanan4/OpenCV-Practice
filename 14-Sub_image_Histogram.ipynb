```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
rainbow = cv.imread('DATA/rainbow.jpg')
show_rainbow = cv.cvtColor(rainbow, cv.COLOR_BGR2RGB)
```

```python
# creating a mask
mask = np.zeros(rainbow.shape[:2], np.uint8)
```

```python
mask[100:400, :100] = 255
```

```python
masked_img = cv.bitwise_or(show_rainbow, show_rainbow, mask = mask)
```

```python
hist_val = cv.calcHist([rainbow], [2], mask, [256], [0,256])
```

```python
plt.imshow(masked_img)
```

```python
plt.plot(hist_val, color = 'r')
```
