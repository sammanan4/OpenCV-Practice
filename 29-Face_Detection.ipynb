<h3> Face detection </h3>
Using pre-trained cascade xml files for various detections (provided by OpenCv).<br>
We create a classifier object using those cascade files.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import the images
nadia = cv.imread('DATA/Nadia_Murad.jpg',0)
denis = cv.imread('DATA/Denis_Mukwege.jpg',0)
solvay = cv.imread('DATA/solvay_conference.jpg',0)
me = cv.imread('E:\\Photos\\Camera\\IMG_20170908_193404.jpg',0)
```

```python
plt.imshow(solvay, cmap='gray')
```

```python
# create the detector
face_detector = cv.CascadeClassifier('DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def findFace(img):
    # create copy of the image
    copy_img = img.copy()
    
    faces = face_detector.detectMultiScale(copy_img)
    
    for x,y,w,h in faces:
        copy_img = cv.rectangle(copy_img, (x,y), (x+w,y+h), (255), 10)
        
    return copy_img
```

```python
result = findFace(nadia)
plt.imshow(result, cmap = 'gray')
```

```python
result = findFace(denis)
```

```python
plt.imshow(result)
```

```python
# detects more faces than there are in the picture
# need to provide more parameters to detectMultiScale method

result = findFace(solvay)
plt.imshow(result, cmap = 'gray')
```

```python
def findFaceAjusted(img):
    # create copy of the image
    copy_img = img.copy()
    count = 0
    
    faces = face_detector.detectMultiScale(copy_img, scaleFactor = 1.13, minNeighbors = 5)
    
    for x,y,w,h in faces:
        count+=1
        copy_img = cv.rectangle(copy_img, (x,y), (x+w,y+h), (255), 10)
        
    return (count, copy_img)
```

```python
nFaces, faces = findFaceAjusted(solvay)
plt.imshow(faces, cmap='gray')
print(nFaces)
```

```python
nFaces, faces = findFaceAjusted(me)
plt.imshow(faces, cmap='gray')
print(nFaces)
```

<h3> Eye detection </h3>

```python
eye_detector = cv.CascadeClassifier('DATA/haarcascades/haarcascade_eye_tree_eyeglasses.xml')
# eye_detector = cv.CascadeClassifier('DATA/haarcascades/haarcascade_eye.xml')
```

```python
def findEyes(img):
    
    img_copy = img.copy()
    
    eyes = eye_detector.detectMultiScale(img_copy)
    
    for x,y,w,h in eyes:
        img_copy = cv.rectangle(img_copy, (x, y), (x+w, y+h), 255, 10)
        
    return img_copy
```

```python
result = findEyes(nadia)
plt.imshow(result, cmap='gray')
```

```python
result = findEyes(denis)
plt.imshow(result, cmap='gray')
```

```python
result = findEyes(me)
plt.imshow(result, cmap='gray')
```

```python
result = findEyes(solvay)
plt.imshow(result, cmap='gray')
```
